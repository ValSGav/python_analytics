#### 2.1. Изучите распределение домов от наличия вида на набережную
#### Постройте график
#### Сделайте выводы

```python
plt.figure(figsize=(8, 6));
data = df_houses.apply(lambda x: "waterfront" if x['waterfront'] > 0 else 'no', axis=1).value_counts(normalize='index');
plt.pie(data, labels=data.index,  autopct='%1.2f%%');
plt.title('Распределение домов по наличию вида на набережную');
```

#### 2.2. Изучите распределение этажей домов

```python
plt.figure(figsize=(8, 6));
data = df_houses['floors'].value_counts(normalize='index');
plt.pie(data, labels=data.index,  autopct='%1.2f%%');
plt.title('Распределение домов по этажности');
```

#### 2.3. Изучите распределение состояния домов

```python
plt.figure(figsize=(8, 6));
def get_condition(condition:int) -> str:
    if condition ==1:
        return "Отвратительное"
    elif condition == 2:
        return "Плохое"
    elif condition == 3:
        return "Среднее"
    elif condition == 4:
        return "Хорошее"
    else:
        return "Отличное"

data = df_houses.apply(lambda x: get_condition(x['condition']), axis=1).value_counts(normalize='index');
plt.bar(data.index, data);
plt.title('Распределение домов по состоянию');
plt.ylabel('Процент домов')
plt.xlabel('Состояние')
```

### Наш продукт поможет определить оптимальную рыночную цену продажи недвижимости в зависимости от разных параметров. В нашей базе более 21 тысячи записей, что позволяет с высокой точностью делать выводы.
#### 1. Например, мы можем проанализировать какие показатели больше всего влияют на цену:

```python
corr_m = df_houses.corr().round(1)
corr_m[np.abs(corr_m)<0.3] = 0
```

```python
sns.heatmap(corr_m)
```

#### На графике представлена корреляция всех показателей между собой. Если анализировать показатель price, видно, что больше всего на него влияют 'sqft_living' , 'grade' , 'bathrooms', 'view'

#### 2. Посмотрим каков разброс цен в зависимости от вида из окон:

```python
plt.figure(figsize = (28, 8));
sns.boxplot(x = df_houses['view'], y = df_houses['price'], showfliers = False)
plt.title('Уровень цен на дома в зависимости от вида из окон')
plt.yticks(range(75000, 4000000, 300000))
plt.xlabel('Вид из окон, где 1 - отвратительный, а 4 - великолепный')
plt.ylabel('Цена в миллионах')
plt.grid()
```

#### Можно сделать вывод, что чем лучше вид из окон, тем выше цена дома. 

#### 3. Посмотрим какая в среднем цена дома в зависимости от количества ванных комнат

```python
plt.figure(figsize = (8, 8));
data = df_houses.groupby('bathrooms')['price'].agg('median');
plt.bar(data.index, data, width=.2);
plt.xlabel('Количество ванных комнат');
plt.ylabel('Цена дома в миллионах');
plt.xticks(range(0,8,1))
plt.title('Зависимость цены дома от количетва ванных комнат');
```

#### Видим, что чем большее количество ванных комнат, тем выше цена

#### 4. Разброс цен в зависимости от оценки

```python
plt.figure(figsize = (12, 8));
plt.scatter(df_houses['grade'], df_houses['price'])
plt.grid()
plt.xlabel('Оценка дома');
plt.ylabel('Цена дома в миллионах');
```

#### Как и в остальных случаях, чем выше оценка дома, тем выше цена. Если оценка выше 6 цены растутут быстрее.

#### 5. Оценим стоимость дома в зависимости от жилой площади:

```python
plt.figure(figsize = (12, 8));
plt.scatter(df_houses['sqft_living'], df_houses['price'])
plt.grid()
plt.xlabel('Жилая площадь в квадратных футах');
plt.ylabel('Цена дома в миллионах');

```

#### Видим, что в основном площади домов от 2000 до 5000 кв. фт., цены в до 2 млн. при дальнейшем увеличении цены растут пропорционально увеличению площади.
