```python
plt.figure(figsize=(12, 8))
plt.hist(df['Memory_Amount'], bins=50)
plt.title('Распределение памяти')
plt.xlabel('GB')
plt.ylabel('Количество')
plt.xticks(range(0, 2050, 128), rotation=30)
plt.grid()
```

```python
plt.figure(figsize=(12, 8))

plt.hist(df['Price_euros'], bins=50)

plt.title('Распределение цены ноутбуков')
plt.xlabel('GB')
plt.ylabel('Количество')
plt.xticks(range(0, 7000, 200), rotation=30)
plt.grid()
```

```python
plt.figure(figsize=(12, 8))

plt.hist(df['Weight'], bins=50)

plt.title('Распределение веса')
plt.xlabel('GB')
plt.ylabel('Количество')
plt.xticks(range(0, 5, 1), rotation=30)
plt.grid()
```

```python
data = pd.crosstab(index=df['Company'], columns=df['Cpu_Company']);
data
```

```python
data = data.drop(columns=['Samsung']).reset_index()
data
```

```python
plt.figure(figsize=(12,6));
n_ticks = np.arange(len(data.Company))
offset = 0.2
width = 0.4
plt.bar(n_ticks - offset, data['AMD'], width=width)
plt.bar(n_ticks + offset, data['Intel'], width=width)


plt.xticks(n_ticks, data['Company'], rotation=50);
plt.grid()
plt.legend(['AMD', 'Intel'])
```

```python
data = pd.crosstab(index=df['Company'], columns=df['Cpu_Company'], normalize='index');
data = data.drop(columns=['Samsung']).reset_index()
```

```python
data
```

```python
plt.figure(figsize=(12,6));

plt.bar(data['Company'], data['AMD'])
plt.bar(data['Company'], data['Intel'], bottom=data['AMD'])


plt.xticks(n_ticks, data['Company'], rotation=50);
plt.grid()
plt.legend(['AMD', 'Intel'])
```
