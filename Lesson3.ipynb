```python
import pandas as pd
```

```python
df = pd.read_csv("laptop_price.csv", encoding="windows-1251")
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
```

```python
df.sample(10)
```

```python
def calculate_mem(stri):
    arr_stri = stri.split()
    gb = 0;
    
    for s in arr_stri:
        if 'GB' in s:
            gb+= float(s[:-2])
        if 'TB' in s:
            gb+= float(s[:-2])*1024
    return int(gb)
```

```python
df['Memory_Amount'] = df['Memory'].apply(calculate_mem)
```

```python
df.sample(10)
```

```python
def get_type_mem(st):
    st = st.split()
    ret_st = ''
    for s in st:
        if s.isalpha():
            if ret_st:
                ret_st+=' '
            ret_st+=s;
    return ret_st
```

```python
df['Memory_Type'] = df['Memory'].apply(get_type_mem)
```

```python
df.sample()
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.sample()
```

```python
df.loc[df['Memory_Type'].str.contains('SSD'), 'SSD'] = 1
```

```python
df.sample()
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float')
```
